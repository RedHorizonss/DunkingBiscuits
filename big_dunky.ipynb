```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
big_df = pd.read_csv('dunking-data.csv')
big_df.head()
```

```python
def plot_L_over_time(df, spacing, name):
    mean_values_over_t = df.groupby(pd.cut(df['t'], spacing))[['t', 'L_squared']].mean()
    plt.scatter(mean_values_over_t['t'], mean_values_over_t['L_squared'], label = name)
```

```python
big_df['L_squared'] = big_df['L']**2
```

```python
dig_df = big_df[big_df['biscuit'] == 'Digestive']
hob_df = big_df[big_df['biscuit'] == 'Hobnob']
rt_df = big_df[big_df['biscuit'] == 'Rich Tea']
```

```python
spacing = np.linspace(10.0, 30.0, num=11)
spacing
```

```python
plot_L_over_time(dig_df, spacing, 'Digestive')
plot_L_over_time(hob_df, spacing, 'Hobnob')
plot_L_over_time(rt_df, spacing, 'Rich Tea')

plt.xlabel('Time (s)')
plt.ylabel(r'$L^2$')
plt.legend()
```

```python
tr1 = pd.read_csv('tr-1.csv')
tr2 = pd.read_csv('tr-2.csv')
tr3 = pd.read_csv('tr-3.csv')

tr1['L_squared'] = tr1['L']**2
tr2['L_squared'] = tr2['L']**2
tr3['L_squared'] = tr3['L']**2

plt.scatter(tr1['t'], tr1['L_squared'], label = 'tr1')
plt.scatter(tr2['t'], tr2['L_squared'], label = 'tr2')
plt.scatter(tr3['t'], tr3['L_squared'], label = 'tr3')

plt.xlabel('Time (s)')
plt.ylabel(r'$L^2$')
plt.legend()
```

```python
plt.scatter(tr1['t'], tr1['L_squared'], label = 'tr1')
plt.scatter(tr2['t'], tr2['L_squared'], label = 'tr2')
plt.scatter(tr3['t'], tr3['L_squared'], label = 'tr3')

plot_L_over_time(dig_df, spacing, 'Digestive')
plot_L_over_time(hob_df, spacing, 'Hobnob')
plot_L_over_time(rt_df, spacing, 'Rich Tea')

plt.xlabel('Time (s)')
plt.ylabel(r'$L^2$')
plt.legend()

plt.show()
```
